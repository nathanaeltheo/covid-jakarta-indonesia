```python
# Importation
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# Data
data = pd.read_csv('C:/Users/mnauf/Harits/Coding/covid_19_indonesia_time_series_all.csv')
data.head()
```

```python
#Choose only Jakarta
df=data.loc[data['Location'] == 'DKI Jakarta']
```

```python
# load dataset for New Cases
val_new_cases = df['New Cases'].values
```

```python
#Differencing to eliminate auto-correlation for New Cases

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

# Original Series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(val_new_cases); axes[0, 0].set_title('Original Series')
plot_acf(val_new_cases, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(np.diff(val_new_cases)); axes[1, 0].set_title('1st Order Differencing')
plot_acf(np.diff(val_new_cases), ax=axes[1, 1])
diffval_new_cases = np.diff(val_new_cases)

# 2nd Differencing
axes[2, 0].plot(np.diff(diffval_new_cases)); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(np.diff(diffval_new_cases), ax=axes[2, 1])
plt.show()
```

```python
#Running ARIMA model for New Cases

from statsmodels.tsa.arima_model import ARIMA
import pmdarima as pm
model_new_cases = pm.auto_arima(val_new_cases, start_p=1, start_q=1,
test='adf', # use adftest to find optimal 'd'
max_p=3, max_q=3, # maximum p and q
m=1, # frequency of series
d=None, # let model determine 'd'
seasonal=False, # No Seasonality
start_P=0,
D=0,
trace=True,
error_action='ignore',
suppress_warnings=True,
stepwise=True)
print(model_new_cases.summary())
```

```python
#Checking statistical features of the model

model_new_cases.plot_diagnostics(figsize=(7,5))
plt.show()
```

```python
# Forecast
n_periods = 50
fc, confint = model_new_cases.predict(n_periods=n_periods, return_conf_int=True)
index_of_fc = np.arange(len(val_new_cases), len(val_new_cases)+n_periods)

# make series for plotting purpose
fc_series = pd.Series(fc, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)

# Plot
plt.plot(val_new_cases)
plt.plot(fc_series, color='darkgreen')
plt.fill_between(lower_series.index,
lower_series,
upper_series,
color='k', alpha=.15)
plt.title("Forecast of New Cases Jakarta")
plt.show()
```
